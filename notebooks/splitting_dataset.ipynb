```python
import os
import shutil
from sklearn.model_selection import train_test_split

SOURCE_DIR = r"C:\Users\ADMIN\Desktop\Projects\Cancer Detection\Esophageal-Cancer-Detection\dataset\raw"
OUTPUT_DIR = r"C:\Users\ADMIN\Desktop\Projects\Cancer Detection\Esophageal-Cancer-Detection\dataset\classification"

def split_class(class_name):

    class_path = os.path.join(SOURCE_DIR, class_name)

    images = os.listdir(class_path)

    # 80% train, 20% test
    train_imgs, test_imgs = train_test_split(
        images,
        test_size=0.20,
        random_state=42,
        shuffle=True
    )

    # 20% of train becomes validation
    train_imgs, val_imgs = train_test_split(
        train_imgs,
        test_size=0.20,
        random_state=42,
        shuffle=True
    )

    splits = {
        "train": train_imgs,
        "val": val_imgs,
        "test": test_imgs
    }

    for split_name, files in splits.items():

        save_dir = os.path.join(
            OUTPUT_DIR,
            split_name,
            class_name
        )

        os.makedirs(save_dir, exist_ok=True)

        for file in files:

            src = os.path.join(class_path, file)
            dst = os.path.join(save_dir, file)

            shutil.copy2(src, dst)

    print(f"\n{class_name}")
    print(f"Train : {len(train_imgs)}")
    print(f"Val   : {len(val_imgs)}")
    print(f"Test  : {len(test_imgs)}")

split_class("cancerous")
split_class("non-Cancerous")

print("\nDataset split completed!")
```
